### Grouping with Functions

```python
# Define a function to group by age range
def age_range(age):
    if isinstance(age, str):
        return 'Unknown'
    age = int(age)
    if age < 30:
        return 'Under 30'
    elif age >= 30 and age < 35:
        return '30-34'
    elif age >= 35 and age < 40:
        return '34-39'
    elif age >= 40 and age < 45:
        return '40-44'
    elif age >= 45 and age < 50:
        return '44-49'
    else:
        return '50+'
```

```python
# Generate age ranges
candy_df_first_6_clean['AGE_RANGE'] = candy_df_first_6_clean['Q3_AGE'].apply(age_range)

candy_df_first_6_clean.head()
```

```python
# Group the dataset using the age_range function
grouped_age_df = candy_df_first_6_clean.groupby('AGE_RANGE')
```

```python
# Iterate over the groups and print the data
for group, data in grouped_age_df:
    print(f"Group: {group}")
    print(data)
    print()
```

### Frequencies and date offsets

```python
import pandas as pd

# Convert 'Timestamp' column to datetime
candy_hier_2015_df['Timestamp'] = pd.to_datetime(candy_hier_2015_df['Timestamp'])
```
